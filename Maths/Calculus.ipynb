# Calculus
Calculus is about how one quantity moves when another changes. Almost every ML training algorithm is calculus in disguise.

       - Derivatives measure how a function's output changes as the input changes.
       - Gradients extend the derivative to many variables — the direction of steepest increase.
       - Gradient descent uses gradients to nudge a model's parameters toward better predictions.
       - The chain rule is what makes backpropagation in neural networks possible.

## Function
A **function** takes an input and produces an output. Written `f(x) = ...`. In ML, the model is a function — feature values go in, predictions come out.

`f(x) = x²` — squares the input
`f(x) = 2x + 3` — linear function

```python
import numpy as np

def f(x):
    return x ** 2

print("f(3) =", f(3))   # 9
print("f(5) =", f(5))   # 25
```

## Limits
The value a function **approaches** as the input gets closer and closer to some target. Written:

`lim (x → a) f(x) = L`

Read as: "as x gets very close to a, f(x) gets very close to L."

Limits are the foundation underneath derivatives — the derivative is the limit of a slope as the gap between two points shrinks to zero.


```python
# f(x) = x² near x = 2
# As x approaches 2, f(x) approaches 4

for x in [1.9, 1.99, 1.999, 2.001, 2.01, 2.1]:
    print(f"f({x}) = {x**2}")
```

## Derivatives
The **derivative** of a function tells you its **rate of change** — how fast `f(x)` is changing with respect to `x` at a given point. It's the slope of the tangent line at that point.

`f'(x) = lim (h → 0)  [ f(x + h) − f(x) ] / h`

Common notation: `f'(x)`, `df/dx`, or `d/dx [f(x)]`.

- `f'(x) > 0` — function is going up at x
- `f'(x) < 0` — function is going down at x
- `f'(x) = 0` — flat (a peak, valley, or saddle point)

**ML use:** the entire training process. The loss function is some `L(weights)`, and we use its derivative to decide how to adjust the weights.

```python
# Numerical derivative of f(x) = x² at x = 3
# The formula says: take h very small, then (f(x+h) - f(x)) / h

def f(x):
    return x ** 2

x = 3
h = 0.0001

derivative = (f(x + h) - f(x)) / h
print(f"f'(3) ≈ {derivative:.4f}")   # should be close to 6 (because d/dx [x²] = 2x)
```

### Common Derivative Rules

A handful of rules cover almost everything you'll meet:

| Function | Derivative |
|---|---|
| `c` (constant) | `0` |
| `x` | `1` |
| `x²` | `2x` |
| `xⁿ` | `n · xⁿ⁻¹` (power rule) |
| `eˣ` | `eˣ` |
| `ln(x)` | `1 / x` |
| `f + g` | `f' + g'` (sum rule) |
| `c · f` | `c · f'` (constant multiple) |
| `f · g` | `f' · g + f · g'` (product rule) |
| `f(g(x))` | `f'(g(x)) · g'(x)` (chain rule — see below) |

**ML use:** these get you the derivative of any model you'll meet for a while. Sigmoid, ReLU, MSE loss — all combinations of the rules above.

```python
# Verify the power rule: d/dx [x³] = 3x²
# At x = 4, this should be 3 * 16 = 48

def f(x):
    return x ** 3

x = 4
h = 0.0001
derivative = (f(x + h) - f(x)) / h

print(f"Numerical: {derivative:.4f}")
print(f"Formula  : {3 * x**2}")
```

## Partial Derivatives
When a function has more than one input, like `f(x, y) = x² + 3y`, you can take the derivative with respect to **one variable at a time**, treating the others as constants.

The derivative of `f` with respect to `x` is written `∂f/∂x` (read "partial f, partial x").

For `f(x, y) = x² + 3y`:
- `∂f/∂x = 2x`  (treat `y` like a constant — its derivative is 0)
- `∂f/∂y = 3`   (treat `x` like a constant)

**ML use:** every model has many parameters (think: thousands of neural network weights). Each one needs its own partial derivative to know how to update it.

```python
# Numerical partial derivatives of f(x, y) = x² + 3y at (x=2, y=5)

def f(x, y):
    return x ** 2 + 3 * y

x, y = 2, 5
h = 0.0001

# Partial with respect to x — change x, keep y the same
df_dx = (f(x + h, y) - f(x, y)) / h

# Partial with respect to y — change y, keep x the same
df_dy = (f(x, y + h) - f(x, y)) / h

print(f"∂f/∂x at (2, 5): {df_dx:.4f}    (formula says 2x = 4)")
print(f"∂f/∂y at (2, 5): {df_dy:.4f}    (formula says 3)")
```

## Gradient
The **gradient** of a multi-variable function is a **vector** holding all its partial derivatives. Written `∇f` (read "del f" or "grad f").

For `f(x, y) = x² + 3y`:

`∇f = [ ∂f/∂x , ∂f/∂y ] = [ 2x , 3 ]`

The gradient points in the **direction of steepest increase**. If you stand at any point on the function's surface and want to climb the fastest, walk in the direction of `∇f`. To go down fastest, walk in the direction of `−∇f`.

**ML use:** training a model means **minimizing a loss function**. The gradient tells us which way is "downhill" in parameter space — and gradient descent uses it to walk in that direction.

```python
# Gradient of f(x, y) = x² + 3y at (x=2, y=5)

def f(x, y):
    return x ** 2 + 3 * y

x, y = 2, 5
h = 0.0001

df_dx = (f(x + h, y) - f(x, y)) / h
df_dy = (f(x, y + h) - f(x, y)) / h

gradient = np.array([df_dx, df_dy])
print(f"∇f at (2, 5) = {gradient.round(4)}")
print(f"This vector points in the direction of steepest increase.")
```

## Gradient Descent
The algorithm that powers almost every ML model. To find the minimum of a function:

1. Start at a random point.
2. Compute the gradient at that point.
3. Take a small step in the **opposite direction** of the gradient (downhill).
4. Repeat until you stop moving much.

The update rule:

`new_x = old_x − learning_rate × gradient`

| Term | Meaning |
|---|---|
| **Learning rate** | How big each step is. Too big → overshoots. Too small → takes forever. |
| **Convergence** | Reaching a point where the gradient is near zero (a minimum). |

**ML use:** every parameter in every trained model — linear regression, logistic regression, neural networks — was set by gradient descent (or a fancier version of it).

```python
# Find the minimum of f(x) = (x - 5)²  using gradient descent
# The true minimum is at x = 5 (where f = 0)
# The derivative is f'(x) = 2(x - 5)

x = 0.0              # starting point
learning_rate = 0.1
steps = 20

for i in range(steps):
    gradient = 2 * (x - 5)            # derivative at current x
    x = x - learning_rate * gradient  # step downhill
    print(f"Step {i+1:2d}: x = {x:.4f}")

print(f"\nFinal x: {x:.4f}  (true minimum is 5)")
```

## Chain Rule
When a function is **made of other functions** — like `h(x) = f(g(x))` — the chain rule tells you how to take its derivative:

`h'(x) = f'(g(x)) · g'(x)`

In words: derivative of the **outer** function (with the inner one still inside), times the derivative of the **inner** function.

Example: `h(x) = (3x + 1)²`
- Inner: `g(x) = 3x + 1`, so `g'(x) = 3`
- Outer: `f(u) = u²`, so `f'(u) = 2u`
- Chain rule: `h'(x) = 2(3x + 1) · 3 = 6(3x + 1)`

**ML use:** this rule is the entire engine of **backpropagation** in neural networks. A neural network is a giant chain of functions; the chain rule is how gradients flow backwards through all the layers to reach each weight.

```python
# Verify the chain rule: h(x) = (3x + 1)²  →  h'(x) = 6(3x + 1)
# At x = 2: h'(2) = 6 * (6 + 1) = 42

def h(x):
    return (3 * x + 1) ** 2

x = 2
h_step = 0.0001

# Numerical derivative
numerical = (h(x + h_step) - h(x)) / h_step

# Chain rule formula
formula = 6 * (3 * x + 1)

print(f"Numerical: {numerical:.4f}")
print(f"Formula  : {formula}")
```

## Integrals
The **opposite** of a derivative. While the derivative measures rate of change, the **integral** measures **accumulated total** — the area under a function's curve.

Written `∫ f(x) dx`. The result is itself a function (the antiderivative).

A **definite integral** between `a` and `b` gives a single number — the exact area under the curve from `a` to `b`:

`∫[a to b] f(x) dx = area under f from x=a to x=b`

**ML use:** less common day-to-day than derivatives. Shows up in probability density functions (the area under a PDF gives a probability), continuous distributions, and some advanced loss functions. You'll usually let libraries handle these — knowing what an integral *means* (accumulated area) is enough for most ML work.

```python
# Approximate the integral of f(x) = x²  from 0 to 3
# True answer:  x³/3  evaluated from 0 to 3  =  27/3 = 9.0

def f(x):
    return x ** 2

# Slice the range into many small rectangles and sum their areas
x_values = np.linspace(0, 3, 1000)
dx = x_values[1] - x_values[0]

area = np.sum(f(x_values) * dx)
print(f"Approximate area under x² from 0 to 3: {area:.4f}")
print(f"True value: 9.0")
```
