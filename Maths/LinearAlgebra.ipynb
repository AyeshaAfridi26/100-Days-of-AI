# Linear Algebra 
 Most datasets and models are represented using vectors and matrices

       - Data in ML is represented as vectors (features) and matrices (datasets).
       - Operations like dot product, matrix multiplication and transformations power ML algorithms.
       - Key concepts such as eigenvalues, eigenvectors and decompositions simplify dimensionality reduction, optimization and training.

## Scalar
Scalars are single numerical values, without direction, magnitude only.

## Vectors
Vectors are quantities that have both magnitude and direction.
Each row in dataframe is a vector
## Matrices
Matrices are rectangular arrays of numbers, arranged in rows and columns. 

````python
loss = 2.45 # Scalar
v = [2,5,6] # Vector
m = [[1,2,3],[4,5,6],[7,8,9]] # Matrix
````

## Core Vector Operations
### Addition and Subtraction
u = [ 2, -1, 4]
v = [ 3, 0, -2]
u +v = [ 5, -1, 2]
u −v = [ -1, -1, 6]

````python
import numpy as np

u = np.array([2, -1, 4])
v = np.array([3, 0, -2])

print(u + v)   # [ 5, -1,  2]
print(u - v)   # [-1, -1,  6]
````

### Scalar Multiplication
Scale every element of a vector by a single number.

`3 × [2, -1, 4] = [6, -3, 12]`

**ML use:** adjusting learning rates, feature scaling, weight updates.

````python
scalar = 3
u = np.array([2, -1, 4])

result = scalar * u
print(f"3 * {u} = {result}")
````

### Dot Product
Sum of element-wise products — produces a **scalar**.

`u · v = u₀v₀ + u₁v₁ + ... + uₙvₙ`

**ML use:** weighted sums in neurons, similarity scoring, the core of matrix multiplication.

````python
u = np.array([2, -1, 4])
v = np.array([3, 0, -2])

dot = np.dot(u, v)
print(f"u · v = {dot}")  # 2*3 + (-1)*0 + 4*(-2) = 6 + 0 - 8 = -2
````

### Vector Magnitude (Norm)
The Euclidean length of a vector — how "big" it is.

`||v|| = √(v₀² + v₁² + ... + vₙ²)`

**ML use:** normalizing vectors, L2 regularization, distance metrics.

````python
u = np.array([2, -1, 4])

magnitude = np.linalg.norm(u)
print(f"||u|| = {magnitude:.4f}")  # sqrt(4 + 1 + 16) = sqrt(21)
````

### Cosine Similarity
Measures the **angle** between two vectors — ignores magnitude, captures direction.

`cos(θ) = (u · v) / (||u|| × ||v||)`

| Value | Meaning |
|-------|---------|
| 1 | Same direction |
| 0 | Perpendicular (unrelated) |
| -1 | Opposite direction |

**ML use:** text/document similarity, recommendation systems, comparing word embeddings.

````python
u = np.array([2, -1, 4])
v = np.array([3, 0, -2])

cos_sim = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
print(f"Cosine Similarity = {cos_sim:.4f}")
# 1 = same direction, 0 = unrelated, -1 = opposite
````

## Matrix Operations

| Operation | Description |
|-----------|-------------|
| `A + B` | Element-wise addition (same shape required) |
| `A - B` | Element-wise subtraction |
| `c * A` | Scalar multiplication |
| `A @ B` | Matrix multiplication — dot product of rows × columns |
| `A.T` | Transpose — flip rows and columns |

**ML use:** forward pass in neural networks (`X @ W + b`), covariance matrices, data transformations.

````python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("A + B:\n", A + B)
print("A - B:\n", A - B)
print("2 * A:\n", 2 * A)
print("A @ B (matrix multiply):\n", A @ B)
print("A.T (transpose):\n", A.T)
````

### Transpose
Flip a matrix over its diagonal — **rows become columns**, columns become rows. A matrix of shape `(m, n)` becomes `(n, m)`.

```
A = [[1, 2, 3],      A.T = [[1, 4],
     [4, 5, 6]]             [2, 5],
                            [3, 6]]
```

**ML use:** aligning dimensions for matrix multiplication (`X.T @ X`), converting row vectors to column vectors, computing covariance matrices.

````python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

print("A shape:", A.shape)
print("A:\n", A)
print("\nA.T shape:", A.T.shape)
print("A.T:\n", A.T)
````

### Identity Matrix
A square matrix with **1s on the diagonal** and 0s everywhere else. Multiplying any matrix by I leaves it unchanged — the matrix equivalent of multiplying by 1.

```
I = [[1, 0, 0],
     [0, 1, 0],
     [0, 0, 1]]

A @ I = I @ A = A
```

**ML use:** weight initialization, ridge regression regularization term `(XᵀX + λI)⁻¹`, theoretical proofs.

````python
I3 = np.eye(3)
print("3×3 Identity Matrix:\n", I3)

A = np.array([[1, 2], [3, 4]])
print("\nA @ I == A:", np.allclose(A @ np.eye(2), A))
````

### Matrix Inverse
The inverse **A⁻¹** undoes the transformation of **A**. Only square, non-singular matrices have one.

`A @ A⁻¹ = A⁻¹ @ A = I`

> A matrix has no inverse if its determinant is 0 (singular matrix — rows are linearly dependent).

**ML use:** solving linear systems `Ax = b → x = A⁻¹b`; the normal equation for linear regression `w = (XᵀX)⁻¹Xᵀy`.

````python
A = np.array([[1., 2.], [3., 4.]])

A_inv = np.linalg.inv(A)
print("A:\n", A)
print("\nA inverse:\n", A_inv)
print("\nA @ A_inv (should be Identity):\n", np.round(A @ A_inv))
````

### Eigenvalues and Eigenvectors
For a square matrix **A**, an eigenvector **v** is a vector that only gets **scaled** (never rotated) when multiplied by A. The scale factor is the eigenvalue **λ**.

`A · v = λ · v`

| Term | Meaning |
|------|---------|
| Eigenvector | The direction preserved under the transformation |
| Eigenvalue | How much that direction is stretched (λ > 1), compressed (0 < λ < 1), or flipped (λ < 0) |

**ML use:** PCA — eigenvectors are the principal components, eigenvalues tell you how much variance each component captures. Also used in spectral clustering and graph analysis.

````python
A = np.array([[2, 1],
              [1, 3]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors (each column is one eigenvector):\n", eigenvectors)

# Verify: A @ v = λ * v
v = eigenvectors[:, 0]
lam = eigenvalues[0]
print(f"\nVerify A·v = λ·v: {np.allclose(A @ v, lam * v)}")
````
