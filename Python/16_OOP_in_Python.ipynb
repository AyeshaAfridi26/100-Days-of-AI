## Part 1: What is OOP and Why Should You Care?

### The two main programming styles

**Procedural programming** is what you've probably been doing: write functions, call them in order, pass data around.

```python
# Procedural style
def calculate_area(width, height):
    return width * height

w, h = 5, 3
print(calculate_area(w, h))
```

**Object-Oriented Programming (OOP)** organizes code into **objects** that bundle data and behavior together.

```python
# OOP style
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

r = Rectangle(5, 3)
print(r.area())
```

### Why use OOP?

| Reason | What it means |
|---|---|
| **Models real-world things** | A `Car`, a `User`, a `BankAccount` — code mirrors how you'd describe these in English |
| **Reusability** | Write a class once, create thousands of objects from it |
| **Organization** | Related data and functions stay together |
| **Maintainability** | Change behavior in one place, affects all objects |
| **Scales to large programs** | Most major software (Django, NumPy, PyTorch) is OOP |

### The core idea: a **class** is a blueprint, an **object** is the actual thing built from it

Think of `class` like an architectural drawing of a house. The drawing is not a house — but you can build many houses from it. Each built house is an **object** (also called an **instance**).

````python
# A quick taste before we dive deep

class Dog:
    def bark(self):
        print("Woof!")

# Create two objects (instances) from the Dog class
dog1 = Dog()
dog2 = Dog()

dog1.bark()
dog2.bark()

# They're separate objects, even though they came from the same class
print(dog1 is dog2)  # False — different objects in memory
````

## Part 3: The `__init__` Constructor and `self`

### `__init__` — the constructor

`__init__` is a special method that runs **automatically when you create an object**. Use it to set up the object's initial attributes.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

### What is `self`?

`self` is **the object itself**. When you write `self.name = name`, you're saying "store this name on the object I'm currently building."

Every method inside a class must have `self` as its first parameter — Python passes it automatically when you call the method.

> **🔑 Key insight:** `self` is just a convention. You could name it anything (like `this` or `obj`), but **always use `self`** — every Python developer expects it.

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def introduce(self):
        print(f"Hi, I'm {self.name} and I'm {self.age} years old.")

# Create objects — arguments go to __init__
alice = Person("Alice", 25)
bob = Person("Bob", 30)

alice.introduce()
bob.introduce()

# Notice: when calling alice.introduce(), Python automatically passes alice as self
````

### What happens under the hood

When you write `alice.introduce()`, Python actually does this:

```python
Person.introduce(alice)
```

It passes `alice` as the `self` argument. That's why the first parameter is always `self` — it receives the object the method was called on.

````python
# Proof: you can call it the long way too
Person.introduce(alice)  # Same as alice.introduce()
````

## Part 4: Attributes and Methods (in depth)

### Two types of attributes

**Instance attributes** belong to a specific object. Each object has its own copy.

**Class attributes** belong to the class itself. ALL objects share the same value.

```python
class Dog:
    species = "Canis familiaris"  # class attribute — shared by all dogs
    
    def __init__(self, name):
        self.name = name           # instance attribute — unique per dog
```

````python
class Dog:
    species = "Canis familiaris"  # class attribute
    
    def __init__(self, name, age):
        self.name = name           # instance attribute
        self.age = age             # instance attribute

rex = Dog("Rex", 5)
buddy = Dog("Buddy", 3)

# Instance attributes are different
print(rex.name, buddy.name)  # Rex Buddy

# Class attribute is shared
print(rex.species)           # Canis familiaris
print(buddy.species)         # Canis familiaris
print(Dog.species)           # Canis familiaris — accessible from the class itself
````

### Common pitfall: mutable class attributes

Using mutable objects (lists, dicts) as class attributes causes bugs because all instances share them.

````python
# BAD — every Dog shares the same tricks list!
class Dog:
    tricks = []  # mutable class attribute
    
    def __init__(self, name):
        self.name = name
    
    def learn_trick(self, trick):
        self.tricks.append(trick)

rex = Dog("Rex")
buddy = Dog("Buddy")

rex.learn_trick("roll over")
print(buddy.tricks)  # ['roll over'] — Buddy didn't learn this!
````

````python
# GOOD — use __init__ for per-instance mutable data
class Dog:
    def __init__(self, name):
        self.name = name
        self.tricks = []  # instance attribute — unique per dog
    
    def learn_trick(self, trick):
        self.tricks.append(trick)

rex = Dog("Rex")
buddy = Dog("Buddy")

rex.learn_trick("roll over")
print("Rex's tricks:", rex.tricks)
print("Buddy's tricks:", buddy.tricks)  # [] — correctly empty
````

### Methods

A **method** is just a function defined inside a class. The first parameter is always `self`.

```python
class Calculator:
    def add(self, a, b):
        return a + b
    
    def multiply(self, a, b):
        return a * b
```

Methods can:
- Read instance attributes (`self.name`)
- Modify instance attributes (`self.score += 10`)
- Call other methods (`self.other_method()`)
- Return values like any function

````python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
    
    def deposit(self, amount):
        if amount <= 0:
            print("Amount must be positive")
            return
        self.balance += amount
        print(f"Deposited ${amount}. New balance: ${self.balance}")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
            return
        self.balance -= amount
        print(f"Withdrew ${amount}. New balance: ${self.balance}")
    
    def display(self):
        print(f"{self.owner}'s account: ${self.balance}")

account = BankAccount("Ahmed", 1000)
account.display()
account.deposit(500)
account.withdraw(200)
account.withdraw(5000)  # Will fail gracefully
````

## Part 5: The Four Pillars of OOP

These four principles are the heart of OOP. 

1. **Encapsulation** — bundling data with the methods that use it; controlling access
2. **Inheritance** — creating new classes based on existing ones
3. **Polymorphism** — different objects responding to the same interface
4. **Abstraction** — hiding complex details behind simple interfaces

Let's go through each one carefully.

### Pillar 1: Encapsulation 

**Encapsulation** means keeping an object's internal data protected from outside interference. The class controls how its data is accessed and modified.

**Why?** If anyone can change an object's data directly, bugs creep in. Encapsulation forces users to go through methods you control.

### Python's "name mangling" conventions

Python doesn't have true `private` or `protected` keywords like Java or C++. It uses naming conventions:

| Prefix | Meaning | Enforced? |
|---|---|---|
| `name` | Public — anyone can access | No restriction |
| `_name` | Protected — "please don't touch" | Convention only |
| `__name` | Private — name-mangled by Python | Sort of (can still be bypassed) |

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner          # public
        self._balance = balance     # protected (convention)
        self.__pin = "1234"         # private (name-mangled)
    
    # Public methods to access private data
    def get_balance(self):
        return self._balance
    
    def verify_pin(self, pin):
        return pin == self.__pin

account = BankAccount("Sara", 5000)
print(account.owner)         # Works fine — public
print(account._balance)      # Works, but you shouldn't access it directly
print(account.get_balance()) # The PROPER way

# Try to access __pin directly — fails!
try:
    print(account.__pin)
except AttributeError as e:
    print(f"Error: {e}")

# Python name-mangles __pin to _BankAccount__pin
print(account._BankAccount__pin)  # Yes, you CAN still access it, but you shouldn't
````

### Better way: properties

Properties let you write methods that look like attributes from the outside, giving you control while keeping clean syntax.

````python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius
    
    @property
    def celsius(self):
        '''Getter — runs when you read t.celsius'''
        return self._celsius
    
    @celsius.setter
    def celsius(self, value):
        '''Setter — runs when you assign t.celsius = ...'''
        if value < -273.15:
            raise ValueError("Temperature below absolute zero!")
        self._celsius = value
    
    @property
    def fahrenheit(self):
        '''Computed property — calculated on the fly'''
        return self._celsius * 9/5 + 32

t = Temperature(25)
print(f"Celsius: {t.celsius}")           # Calls the getter
print(f"Fahrenheit: {t.fahrenheit}")     # Calls the computed property

t.celsius = 100  # Calls the setter — validation runs
print(f"Now: {t.celsius}°C = {t.fahrenheit}°F")

# Validation in action
try:
    t.celsius = -300
except ValueError as e:
    print(f"Caught: {e}")
````

### Pillar 2: Inheritance 

**Inheritance** lets a class take all the attributes and methods of another class, then add or change behavior.

**Why?** You write the common parts once, then build specialized versions on top.

### Terminology
- **Parent class** (also called **base class** or **superclass**): the class being inherited from
- **Child class** (also called **derived class** or **subclass**): the class doing the inheriting

```python
class Parent:
    pass

class Child(Parent):  # Child inherits from Parent
    pass
```

````python
# Base class
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def eat(self):
        print(f"{self.name} is eating.")
    
    def sleep(self):
        print(f"{self.name} is sleeping.")

# Child class — inherits everything from Animal
class Dog(Animal):
    def bark(self):
        print(f"{self.name} says: Woof!")

# Another child class
class Cat(Animal):
    def meow(self):
        print(f"{self.name} says: Meow!")

rex = Dog("Rex", 5)
rex.eat()     # Inherited from Animal
rex.sleep()   # Inherited from Animal
rex.bark()    # Defined in Dog

whiskers = Cat("Whiskers", 3)
whiskers.eat()
whiskers.meow()
````

### Adding to `__init__` in a child class

If a child needs extra attributes, override `__init__` and call the parent's `__init__` with `super()`.

````python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def describe(self):
        print(f"{self.name} is {self.age} years old.")

class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)  # Call Animal's __init__
        self.breed = breed
    
    def describe(self):  # Override the parent method
        super().describe()  # Call the parent version
        print(f"{self.name} is a {self.breed}.")

rex = Dog("Rex", 5, "Golden Retriever")
rex.describe()
````

### Method overriding

A child can replace a parent's method by defining a method with the same name. This is called **overriding**.

````python
class Vehicle:
    def start(self):
        print("Vehicle is starting...")

class Car(Vehicle):
    def start(self):  # Override — replaces parent's start
        print("Car engine roars to life!")

class ElectricCar(Car):
    def start(self):
        print("Electric car silently powers on...")

v = Vehicle()
c = Car()
e = ElectricCar()

v.start()
c.start()
e.start()
````

### Multiple inheritance

A class can inherit from more than one parent. Use cautiously — it can get complicated.

````python
class Swimmer:
    def swim(self):
        print("Swimming...")

class Flyer:
    def fly(self):
        print("Flying...")

class Duck(Swimmer, Flyer):  # Inherits from both
    def quack(self):
        print("Quack!")

donald = Duck()
donald.swim()
donald.fly()
donald.quack()

# Method Resolution Order (MRO) — Python's way of resolving which parent's method to use
print(Duck.__mro__)
````

### Pillar 3: Polymorphism 

**Polymorphism** means "many forms." Different objects can respond to the same method name in their own way.

**Why?** You can write code that works with any object as long as it has the right methods — without caring about the exact type.

This is what makes OOP code so flexible. You'll see why with examples.

````python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        raise NotImplementedError("Subclasses must implement speak()")

class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

class Cow(Animal):
    def speak(self):
        return f"{self.name} says Moo!"

# Polymorphism in action — same code works for all animal types
animals = [Dog("Rex"), Cat("Whiskers"), Cow("Bessie")]

for animal in animals:
    print(animal.speak())  # We don't care what type — they all have speak()
````

### Duck typing — Python's special flavor of polymorphism

> "If it walks like a duck and quacks like a duck, it's a duck."

Python doesn't require objects to share a parent class to be polymorphic. As long as they have the right methods, they work.

````python
class Duck:
    def quack(self):
        return "Quack!"

class Person:
    def quack(self):
        return "I'm pretending to be a duck"

class Robot:
    def quack(self):
        return "BEEP BOOP QUACK"

# None of these inherit from a common class, but they all have quack()
def make_it_quack(thing):
    print(thing.quack())

for obj in [Duck(), Person(), Robot()]:
    make_it_quack(obj)  # Works for all of them!
````

### Pillar 4: Abstraction 

**Abstraction** means hiding complex implementation details behind a simple interface.

**Why?** Users of your class should know *what* it does, not *how* it does it. Think of driving a car: you press the gas pedal (simple interface), and you don't care about fuel injection, combustion timing, etc. (hidden complexity).

### Abstract base classes (ABCs)

Python provides the `abc` module to define abstract classes — classes that can't be instantiated directly and force subclasses to implement specific methods.

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    '''Abstract base class — can't be instantiated directly'''
    
    @abstractmethod
    def area(self):
        '''Subclasses MUST implement this'''
        pass
    
    @abstractmethod
    def perimeter(self):
        '''Subclasses MUST implement this'''
        pass
    
    def describe(self):
        '''Concrete method — inherited as-is'''
        print(f"This shape has area {self.area()} and perimeter {self.perimeter()}")

# Try to instantiate Shape directly — fails!
try:
    s = Shape()
except TypeError as e:
    print(f"Error: {e}")

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14159 * self.radius ** 2
    
    def perimeter(self):
        return 2 * 3.14159 * self.radius

rect = Rectangle(5, 3)
circle = Circle(4)
rect.describe()
circle.describe()
````

````python
# What if a subclass FORGETS to implement an abstract method?
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def area(self):
        return 0.5 * self.base * self.height
    # Forgot to implement perimeter()!

try:
    t = Triangle(3, 4)
except TypeError as e:
    print(f"Error: {e}")
# Python prevents you from creating an incomplete subclass
````

## Part 6: Special "Dunder" Methods

Methods with **d**ouble **under**scores on both sides (like `__init__`) are called **dunder methods** (or "magic methods"). Python calls them automatically in specific situations.

These let your objects work with built-in Python operations: printing, comparing, adding, indexing, etc.

### Most useful dunder methods

| Method | When it's called |
|---|---|
| `__init__(self, ...)` | When the object is created |
| `__str__(self)` | When `str(obj)` or `print(obj)` is called — for humans |
| `__repr__(self)` | When `repr(obj)` is called or in the REPL — for developers |
| `__len__(self)` | When `len(obj)` is called |
| `__eq__(self, other)` | When `obj1 == obj2` is used |
| `__lt__(self, other)` | When `obj1 < obj2` is used |
| `__add__(self, other)` | When `obj1 + obj2` is used |
| `__getitem__(self, key)` | When `obj[key]` is used |
| `__contains__(self, item)` | When `item in obj` is used |
| `__call__(self, ...)` | When you call the object like a function: `obj()` |

````python
# Without dunder methods — default behavior is ugly
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

b = Book("The Hobbit", "Tolkien", 310)
print(b)         # <__main__.Book object at 0x...>  — ugly!
print(b == Book("The Hobbit", "Tolkien", 310))  # False — they're different objects
````

````python
# With dunder methods — clean, intuitive behavior
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    def __str__(self):
        '''Human-readable string — used by print()'''
        return f"'{self.title}' by {self.author}"
    
    def __repr__(self):
        '''Developer-readable string — shows how to recreate'''
        return f"Book({self.title!r}, {self.author!r}, {self.pages})"
    
    def __len__(self):
        '''So len(book) works'''
        return self.pages
    
    def __eq__(self, other):
        '''So book1 == book2 works'''
        if not isinstance(other, Book):
            return False
        return (self.title == other.title and 
                self.author == other.author and 
                self.pages == other.pages)
    
    def __lt__(self, other):
        '''So book1 < book2 works (compares by page count)'''
        return self.pages < other.pages

hobbit = Book("The Hobbit", "Tolkien", 310)
lotr = Book("Lord of the Rings", "Tolkien", 1178)

print(hobbit)                  # 'The Hobbit' by Tolkien
print(repr(hobbit))            # Book('The Hobbit', 'Tolkien', 310)
print(len(hobbit))             # 310
print(hobbit == Book("The Hobbit", "Tolkien", 310))  # True!
print(hobbit < lotr)           # True — Hobbit has fewer pages

# Now we can even sort books!
books = [lotr, hobbit, Book("Mistborn", "Sanderson", 657)]
sorted_books = sorted(books)
for b in sorted_books:
    print(f"{b}: {len(b)} pages")
````

### Operator overloading — making `+` work on your objects

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        '''Vector addition: v1 + v2'''
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        '''Vector subtraction'''
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
        '''Scalar multiplication: v * 3'''
        return Vector(self.x * scalar, self.y * scalar)
    
    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(5, 1)

print(v1 + v2)    # Vector(7, 4)
print(v1 - v2)    # Vector(-3, 2)
print(v1 * 3)     # Vector(6, 9)
````

## Part 7: Class Methods, Static Methods, and Properties

There are three kinds of methods in Python:

| Type | First argument | Purpose |
|---|---|---|
| **Instance method** | `self` | Works on a specific object's data (most common) |
| **Class method** | `cls` (the class itself) | Works on the class — often used as alternative constructors |
| **Static method** | None | Utility function that belongs to the class but doesn't use `self` or `cls` |

````python
class Pizza:
    # Class attribute
    available_sizes = ["small", "medium", "large"]
    
    def __init__(self, size, toppings):
        self.size = size
        self.toppings = toppings
    
    # Instance method — works on self
    def describe(self):
        return f"{self.size} pizza with {', '.join(self.toppings)}"
    
    # Class method — alternative constructor
    @classmethod
    def margherita(cls, size):
        return cls(size, ["mozzarella", "tomato", "basil"])
    
    @classmethod
    def pepperoni(cls, size):
        return cls(size, ["mozzarella", "tomato", "pepperoni"])
    
    # Static method — utility related to pizzas but doesn't need self or cls
    @staticmethod
    def is_valid_size(size):
        return size in Pizza.available_sizes

# Regular creation
p1 = Pizza("large", ["cheese", "mushrooms"])
print(p1.describe())

# Using class methods as alternative constructors
p2 = Pizza.margherita("medium")
print(p2.describe())

p3 = Pizza.pepperoni("small")
print(p3.describe())

# Static method
print(Pizza.is_valid_size("medium"))    # True
print(Pizza.is_valid_size("gigantic"))  # False
````

### When to use which?

- **Instance method (`self`):** when you need to access/modify a specific object's data — the default choice
- **Class method (`@classmethod`):** for alternative constructors or methods that operate on the class itself
- **Static method (`@staticmethod`):** when the method is logically related to the class but doesn't use any object/class data

## Part 8: A Complete Real-World Example

Let's build a small **library management system** that uses everything you've learned:
- Classes and inheritance
- Encapsulation with properties
- Polymorphism through method overriding
- Abstraction with abstract base classes
- Dunder methods for clean syntax

````python
from abc import ABC, abstractmethod
from datetime import datetime, timedelta

# Abstract base class
class LibraryItem(ABC):
    def __init__(self, title, item_id):
        self.title = title
        self.item_id = item_id
        self._is_borrowed = False
        self._borrower = None
        self._due_date = None
    
    @abstractmethod
    def loan_period(self):
        '''Days the item can be borrowed — different per type'''
        pass
    
    @property
    def is_available(self):
        return not self._is_borrowed
    
    def borrow(self, borrower_name):
        if self._is_borrowed:
            return f" '{self.title}' is already borrowed by {self._borrower}"
        self._is_borrowed = True
        self._borrower = borrower_name
        self._due_date = datetime.now() + timedelta(days=self.loan_period())
        return f" '{self.title}' borrowed by {borrower_name}. Due: {self._due_date.strftime('%Y-%m-%d')}"
    
    def return_item(self):
        if not self._is_borrowed:
            return f" '{self.title}' wasn't borrowed."
        borrower = self._borrower
        self._is_borrowed = False
        self._borrower = None
        self._due_date = None
        return f" '{self.title}' returned by {borrower}"
    
    def __str__(self):
        status = "Available" if self.is_available else f"Borrowed by {self._borrower}"
        return f"[{self.item_id}] {self.title} — {status}"


# Concrete subclasses
class Book(LibraryItem):
    def __init__(self, title, item_id, author, pages):
        super().__init__(title, item_id)
        self.author = author
        self.pages = pages
    
    def loan_period(self):
        return 14  # Books: 2 weeks
    
    def __str__(self):
        return f" {super().__str__()} by {self.author}"


class DVD(LibraryItem):
    def __init__(self, title, item_id, director, duration_minutes):
        super().__init__(title, item_id)
        self.director = director
        self.duration = duration_minutes
    
    def loan_period(self):
        return 3  # DVDs: 3 days
    
    def __str__(self):
        return f" {super().__str__()} directed by {self.director}"


class Magazine(LibraryItem):
    def __init__(self, title, item_id, issue):
        super().__init__(title, item_id)
        self.issue = issue
    
    def loan_period(self):
        return 7  # Magazines: 1 week
    
    def __str__(self):
        return f" {super().__str__()} (Issue {self.issue})"


# The library itself
class Library:
    def __init__(self, name):
        self.name = name
        self.items = []
    
    def add_item(self, item):
        self.items.append(item)
        print(f" Added: {item.title}")
    
    def find_by_title(self, title):
        return [item for item in self.items if title.lower() in item.title.lower()]
    
    def available_items(self):
        return [item for item in self.items if item.is_available]
    
    def show_catalog(self):
        print(f"\n {self.name} Catalog:")
        print("-" * 50)
        for item in self.items:
            print(item)
    
    def __len__(self):
        return len(self.items)


# Use it!
library = Library("City Central Library")

library.add_item(Book("The Hobbit", "B001", "J.R.R. Tolkien", 310))
library.add_item(Book("1984", "B002", "George Orwell", 328))
library.add_item(DVD("Inception", "D001", "Christopher Nolan", 148))
library.add_item(Magazine("National Geographic", "M001", "March 2024"))

library.show_catalog()
print(f"\nTotal items: {len(library)}")

print("\n--- Borrowing ---")
print(library.items[0].borrow("Ahmed"))
print(library.items[0].borrow("Sara"))  # Already borrowed
print(library.items[2].borrow("Fatima"))  # DVD

print("\n--- Available items ---")
for item in library.available_items():
    print(item)

print("\n--- Returning ---")
print(library.items[0].return_item())
````

**What this example demonstrates:**

- `LibraryItem` is an abstract base class — you can't create a generic "item," only specific types
- `Book`, `DVD`, `Magazine` all inherit from `LibraryItem` and add their own attributes
- Each subclass implements `loan_period()` differently — **polymorphism**
- `_is_borrowed`, `_borrower`, `_due_date` are protected — **encapsulation**
- `is_available` is a property — clean access to internal state
- `__str__` is overridden so printing items looks nice — **dunder methods**
- `super()` lets child classes call parent methods
- The `Library` class manages many items — **composition**

## Part 9: Common Mistakes and Best Practices

### Mistake 1: Forgetting `self`

```python
class Counter:
    def __init__():           #  Missing self
        count = 0             #  Missing self.

    def increment():          #  Missing self
        count += 1            #  Missing self.
```

**Fix:** Always include `self` as the first parameter, and use `self.attribute` to access object data.

````python
# Correct version
class Counter:
    def __init__(self):
        self.count = 0
    
    def increment(self):
        self.count += 1

c = Counter()
c.increment()
c.increment()
print(c.count)  # 2
````

### Mistake 2: Using mutable defaults in `__init__`

Already shown earlier, but worth repeating because it's so common.

````python
# BAD
class Cart:
    def __init__(self, items=[]):  # Shared list across all instances!
        self.items = items

cart1 = Cart()
cart1.items.append("apple")
cart2 = Cart()
print(cart2.items)  # ['apple'] — shocking!

# GOOD
class Cart:
    def __init__(self, items=None):
        self.items = items if items is not None else []

cart1 = Cart()
cart1.items.append("apple")
cart2 = Cart()
print(cart2.items)  # [] — correct!
````

### Mistake 3: Confusing class vs instance attributes

When you assign to `self.attr`, you create an instance attribute. When you assign to `ClassName.attr`, you change the class attribute.

````python
class Dog:
    species = "Canis familiaris"  # Class attribute
    
    def __init__(self, name):
        self.name = name

rex = Dog("Rex")
buddy = Dog("Buddy")

# This creates an INSTANCE attribute on rex, shadowing the class attribute
rex.species = "Special breed"

print(rex.species)    # Special breed (instance attribute)
print(buddy.species)  # Canis familiaris (class attribute, unchanged)
print(Dog.species)    # Canis familiaris (class attribute)
````

### Best practices checklist

-  Use `PascalCase` for class names (`BankAccount`, not `bank_account`)
-  Use `snake_case` for methods and attributes (`get_balance`, not `getBalance`)
-  Add docstrings to classes and methods (triple-quoted strings)
-  Use `__init__` to set up attributes — don't rely on adding them later
-  Use properties (`@property`) instead of getter/setter methods
-  Prefer composition (one class uses another) over deep inheritance chains
-  Override `__str__` and `__repr__` for any class you'll print or debug
-  Use `super()` to call parent methods — don't hardcode the parent class name
-  Don't make everything a class — sometimes a function is enough

````python
# Exercise 1: Student class

class Student:
    def __init__(self, name, roll_number):
        self.name = name
        self.roll_number = roll_number
        self.grades = {}
    
    def add_grade(self, subject, score):
        self.grades[subject] = score
    
    def average(self):
        if not self.grades:
            return 0
        return sum(self.grades.values()) / len(self.grades)
    
    def __str__(self):
        return f"Student {self.roll_number}: {self.name} — Avg: {self.average():.1f}"

s = Student("Ali", "CS-2024-001")
s.add_grade("Math", 85)
s.add_grade("Science", 92)
s.add_grade("English", 78)
print(s)
````

````python
# Exercise 2: GraduateStudent

class GraduateStudent(Student):
    def __init__(self, name, roll_number, thesis_topic):
        super().__init__(name, roll_number)
        self.thesis_topic = thesis_topic
    
    def __str__(self):
        return super().__str__() + f" — Thesis: {self.thesis_topic}"

g = GraduateStudent("Sara", "MS-2024-007", "ML for Threat Detection")
g.add_grade("Advanced Algorithms", 95)
g.add_grade("Research Methods", 88)
print(g)
````

````python
# Exercise 3: Stack

class Stack:
    def __init__(self):
        self._items = []
    
    def push(self, item):
        self._items.append(item)
    
    def pop(self):
        if not self._items:
            raise IndexError("pop from empty stack")
        return self._items.pop()
    
    def peek(self):
        if not self._items:
            return None
        return self._items[-1]
    
    def __len__(self):
        return len(self._items)
    
    def __bool__(self):
        return len(self._items) > 0
    
    def __str__(self):
        return f"Stack({self._items})"

s = Stack()
s.push(1)
s.push(2)
s.push(3)
print(s)           # Stack([1, 2, 3])
print(s.peek())    # 3
print(s.pop())     # 3
print(len(s))      # 2
print(bool(s))     # True

while s:           # __bool__ makes this work
    print("Popped:", s.pop())
print(bool(s))     # False
````
