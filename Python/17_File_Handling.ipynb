## **1. Introduction to File Handling**

### **What is File Handling?**

File handling is the process of working with files stored on your computer's disk. In Python, you can:
- **Read** data from files
- **Write** data to files
- **Append** data to existing files
- **Delete** or **modify** files

### **Why is File Handling Important?**

1. **Data Persistence**: Store data permanently (survives program termination)
2. **Data Exchange**: Share data between different programs
3. **Large Datasets**: Process data too large to fit in memory
4. **Configuration**: Store settings and preferences
5. **Logging**: Record program activities and errors

### **File Handling Workflow**

```
1. OPEN the file
   â†“
2. PERFORM operations (read/write)
   â†“
3. CLOSE the file
```

**Important**: Always close files after use to:
- Free system resources
- Ensure data is written to disk
- Prevent file corruption
- Allow other programs to access the file

## **3. File Modes Explained**

File modes determine **how** you can interact with a file. Understanding modes is crucial!

### **Basic Modes**

| Mode | Name | Description | Creates File? | Cursor Position |
|------|------|-------------|---------------|------------------|
| `'r'` | Read | Read only (default) |  Error if missing | Start |
| `'w'` | Write | Write only |  Yes | Start (overwrites!) |
| `'a'` | Append | Write only |  Yes | End |
| `'x'` | Exclusive | Write only |  Error if exists | Start |

### **Additional Modes (Combined)**

| Mode | Description | Use Case |
|------|-------------|----------|
| `'r+'` | Read & Write | Modify existing file |
| `'w+'` | Write & Read | Create new, read back |
| `'a+'` | Append & Read | Add to end, read all |

### **Binary Mode Suffix**

Add `'b'` for binary files:
- `'rb'` - Read binary
- `'wb'` - Write binary
- `'ab'` - Append binary

### ** Important Mode Behaviors**

1. **`'w'` mode is destructive!**
   ```python
   # This DELETES all existing content!
   f = open('file.txt', 'w')
   ```

2. **`'r'` requires file to exist**
   ```python
   # This raises FileNotFoundError if file doesn't exist
   f = open('missing.txt', 'r')  #  Error
   ```

3. **`'x'` prevents accidental overwrites**
   ```python
   # Safer than 'w' - won't overwrite existing file
   f = open('important.txt', 'x')  #  Safe
   ```

````python
# Demonstration of different file modes

print("Mode Demonstration\n" + "="*60)

# 1. Write mode 'w' - Creates new or overwrites
print("\n1ï¸ Write mode ('w'):")
with open('demo_write.txt', 'w') as f:
    f.write('First line\n')
    f.write('Second line\n')
print("    File created with 2 lines")

# Writing again - OVERWRITES everything!
with open('demo_write.txt', 'w') as f:
    f.write('This replaces everything!\n')
print("     Previous content was overwritten")

# 2. Append mode 'a' - Adds to end
print("\n2ï¸ Append mode ('a'):")
with open('demo_append.txt', 'w') as f:
    f.write('Line 1\n')

with open('demo_append.txt', 'a') as f:
    f.write('Line 2\n')
    f.write('Line 3\n')

with open('demo_append.txt', 'r') as f:
    print(f"   Content:\n{f.read()}")

# 3. Read mode 'r' - Read only
print("3ï¸ Read mode ('r'):")
with open('demo_append.txt', 'r') as f:
    content = f.read()
    print(f"   Read {len(content)} characters")

# 4. Exclusive mode 'x' - Safe creation
print("\n4ï¸ Exclusive mode ('x'):")
try:
    with open('demo_exclusive.txt', 'x') as f:
        f.write('Created safely\n')
    print("    File created")
    
    # Try again - will fail
    with open('demo_exclusive.txt', 'x') as f:
        f.write('This will fail\n')
except FileExistsError:
    print("    File already exists - prevented overwrite!")

print("\n" + "="*60)
````

## **4. Reading Files**

Python provides multiple methods to read file content, each suited for different scenarios.

### **4.1 The `read()` Method**

**Syntax**: `file.read(size=-1)`

- **No argument**: Reads entire file
- **With size**: Reads specified number of characters
- **Returns**: String (text mode) or bytes (binary mode)

**Use when**: File is small enough to fit in memory

**Memory consideration**: 
- Reading a 1GB file uses 1GB+ of RAM!
- For large files, use `readline()` or iteration

````python
# Create a sample file
with open('reading_demo.txt', 'w') as f:
    f.write('Line 1: Python is awesome!\n')
    f.write('Line 2: File handling is important.\n')
    f.write('Line 3: Practice makes perfect.\n')

print("Reading Methods Demo\n" + "="*60)

# Method 1: read() - Entire file
print("\n1ï¸ read() - Read entire file:")
print("-" * 60)
with open('reading_demo.txt', 'r') as f:
    content = f.read()
    print(content)
    print(f"Type: {type(content)}")
    print(f"Length: {len(content)} characters")

# Method 2: read(size) - Specific number of characters
print("\n2ï¸ read(size) - Read first 20 characters:")
print("-" * 60)
with open('reading_demo.txt', 'r') as f:
    chunk = f.read(20)
    print(f"First 20 chars: '{chunk}'")
    
    # Read next 20
    chunk2 = f.read(20)
    print(f"Next 20 chars: '{chunk2}'")
    print("\n Notice: File cursor moves forward after each read!")
````

### **4.2 The `readline()` Method**

**Syntax**: `file.readline(size=-1)`

- Reads **one line** at a time
- Returns line including newline character (`\n`)
- Returns empty string `''` at end of file

**Use when**: Processing large files line by line

**Memory efficient**: Only one line in memory at a time

````python
print("\n3ï¸ readline() - Read line by line:")
print("-" * 60)

with open('reading_demo.txt', 'r') as f:
    line1 = f.readline()
    print(f"Line 1: {line1}", end='')  # end='' because line has \n
    
    line2 = f.readline()
    print(f"Line 2: {line2}", end='')
    
    line3 = f.readline()
    print(f"Line 3: {line3}", end='')
    
    line4 = f.readline()
    print(f"\nLine 4 (EOF): '{line4}'")
    print("   Empty string means end of file!")

# Using readline() in a loop
print("\n4ï¸ readline() in loop:")
print("-" * 60)

with open('reading_demo.txt', 'r') as f:
    line_num = 1
    while True:
        line = f.readline()
        if not line:  # Empty string = EOF
            break
        print(f"Line {line_num}: {line.strip()}")  # strip() removes \n
        line_num += 1
````

### **4.3 The `readlines()` Method**

**Syntax**: `file.readlines()`

- Reads **all lines** into a **list**
- Each line is a list element (includes `\n`)
- Returns empty list if file is empty

**Use when**: 
- Need to process all lines
- File is small enough for memory
- Need random access to lines

````python
print("\n5ï¸ readlines() - All lines as list:")
print("-" * 60)

with open('reading_demo.txt', 'r') as f:
    lines = f.readlines()

print(f"Type: {type(lines)}")
print(f"Number of lines: {len(lines)}")
print(f"\nLines list: {lines}")

print("\nAccessing individual lines:")
for i, line in enumerate(lines):
    print(f"  lines[{i}] = {repr(line)}")

# Processing the list
print("\nProcessed lines (stripped):")
for line in lines:
    print(f"  â€¢ {line.strip()}")
````

### **4.4 Iterating Over File Object (BEST PRACTICE)**

**Most Pythonic way** to read files:

```python
with open('file.txt', 'r') as f:
    for line in f:
        print(line)
```

**Advantages**:
-  Memory efficient (one line at a time)
-  Clean and readable
-  Automatic cleanup
-  Works with huge files

**This is the recommended approach for most file reading!**

````python
print("\n 6ï¸ File iteration (RECOMMENDED):")
print("-" * 60)

with open('reading_demo.txt', 'r') as f:
    for line_num, line in enumerate(f, 1):
        print(f"{line_num}. {line.strip()}")

print("\n This is the most efficient and Pythonic way!")

# Comparison of methods
print("\n" + "="*60)
print(" Comparison of Reading Methods:")
print("="*60)
print("\nMethod         | Memory Usage | Speed  | Use Case")
print("-" * 60)
print("read()         | High         | Fast   | Small files")
print("readline()     | Low          | Medium | Line-by-line processing")
print("readlines()    | High         | Fast   | Need all lines as list")
print("Iteration      | Low          | Fast   | Most cases (BEST!)")
print("="*60)
````

## **5. Writing Files**

### **5.1 The `write()` Method**

**Syntax**: `file.write(string)`

- Writes a string to the file
- Returns number of characters written
- **Does NOT add newline automatically** - you must add `\n`

**Important**: 
- Data might be buffered (not immediately written to disk)
- Use `flush()` to force write, or close the file

````python
print("Writing Files Demo\n" + "="*60)

# Basic write()
print("\n 1ï¸ Basic write():")
print("-" * 60)

with open('output.txt', 'w') as f:
    chars_written = f.write('Hello, World!')
    print(f"Wrote {chars_written} characters")
    
    # Write more - need to add \n manually
    f.write('\n')  # Add newline
    f.write('This is line 2\n')
    f.write('This is line 3')

# Verify
with open('output.txt', 'r') as f:
    print("\nFile content:")
    print(f.read())

# Writing multiple lines efficiently
print("\n 2ï¸ Writing multiple lines:")
print("-" * 60)

lines_to_write = [
    'First line\n',
    'Second line\n',
    'Third line\n'
]

with open('multi_line.txt', 'w') as f:
    for line in lines_to_write:
        f.write(line)

print(" Multiple lines written")

# Writing numbers and other types
print("\n 3 Writing different data types:")
print("-" * 60)

with open('data.txt', 'w') as f:
    # Must convert to string!
    name = "Alice"
    age = 25
    score = 95.5
    
    f.write(f"Name: {name}\n")
    f.write(f"Age: {age}\n")
    f.write(f"Score: {score}\n")
    
    # This would cause an error:
    # f.write(age)  #   TypeError: expected str, got int

with open('data.txt', 'r') as f:
    print(f.read())
````

### **5.2 The `writelines()` Method**

**Syntax**: `file.writelines(list_of_strings)`

- Writes a list of strings to file
- **Does NOT add newlines** - must include in strings
- More efficient than multiple `write()` calls

**Common mistake**: Forgetting to add `\n` to each string

````python
print("\n 4ï¸ writelines() method:")
print("-" * 60)

# Correct way - with newlines
lines = [
    'Apple\n',
    'Banana\n',
    'Cherry\n',
    'Date\n'
]

with open('fruits.txt', 'w') as f:
    f.writelines(lines)

print("With newlines:")
with open('fruits.txt', 'r') as f:
    print(f.read())

# Without newlines - everything on one line!
lines_no_newline = ['Apple ', 'Banana ', 'Cherry ']

with open('fruits_bad.txt', 'w') as f:
    f.writelines(lines_no_newline)

print("\nWithout newlines:")
with open('fruits_bad.txt', 'r') as f:
    print(f"Result: '{f.read()}'")
    print("     All on one line!")
````

## **7. Context Managers (`with` statement)**

### **Why Use `with`?**

**Problem with manual file handling**:
```python
f = open('file.txt', 'r')
content = f.read()
# If error occurs here, file never closes!
f.close()
```

**Solution - Context Manager**:
```python
with open('file.txt', 'r') as f:
    content = f.read()
# File automatically closed, even if error occurs!
```

### **Benefits of `with`**:

1. **Automatic Cleanup**: File closed even if exception occurs
2. **Cleaner Code**: No need for try-finally
3. **Better Practice**: Reduces bugs from forgotten `close()`
4. **Resource Management**: Prevents resource leaks

````python
print("Context Manager Demo\n" + "="*60)

# BAD: Without context manager
print("\n Without 'with' (Not recommended):")
print("-" * 60)
f = open('test.txt', 'w')
f.write('Some data')
f.close()  # Must remember to close!
print("File written and closed manually")

# GOOD: With context manager
print("\n With 'with' (Recommended):")
print("-" * 60)
with open('test.txt', 'w') as f:
    f.write('Some data')
# Automatically closed here!
print("File automatically closed")

# Demonstrating automatic cleanup
print("\n Proving Automatic Closure:")
print("-" * 60)

with open('auto_close.txt', 'w') as f:
    print(f"Inside 'with': File closed? {f.closed}")
    f.write('test')

print(f"Outside 'with': File closed? {f.closed}")

# Exception handling
print("\n With Exception Handling:")
print("-" * 60)

try:
    with open('exception_test.txt', 'w') as f:
        f.write('Start')
        # Simulate error
        x = 1 / 0  # This will raise ZeroDivisionError
        f.write('End')  # This won't execute
except ZeroDivisionError:
    print("Error occurred!")

print(f"File still closed properly: {f.closed}")
print(" Even with error, file was closed!")

# Multiple files
print("\n Opening Multiple Files:")
print("-" * 60)

with open('source.txt', 'w') as source:
    source.write('Source content')

# Copy file
with open('source.txt', 'r') as source, open('dest.txt', 'w') as dest:
    content = source.read()
    dest.write(content)

print(" File copied using multiple context managers")
````

| Method | Description | Example |
|--------|-------------|----------|

## **9. Binary Files**

### **Text vs Binary Mode**

**Text Mode** (`'r'`, `'w'`, `'a'`):
- Works with strings
- Handles encoding/decoding
- Platform-specific newline conversion
- Use for: `.txt`, `.csv`, `.json`, `.xml`

**Binary Mode** (`'rb'`, `'wb'`, `'ab'`):
- Works with bytes
- No encoding
- No newline conversion
- Use for: images, videos, executables, PDFs

### **When to Use Binary Mode**

- Images (`.png`, `.jpg`, `.gif`)
- Videos (`.mp4`, `.avi`)
- Audio (`.mp3`, `.wav`)
- Archives (`.zip`, `.tar`)
- Executables (`.exe`, `.dll`)
- PDFs (`.pdf`)

````python
print("Binary Files Demo\n" + "="*60)

# Writing binary data
print("\n 1ï¸ Writing Binary Data:")
print("-" * 60)

# Create sample binary data
data = bytes([65, 66, 67, 68, 69])  # ASCII: A B C D E
print(f"Binary data: {data}")
print(f"Type: {type(data)}")

with open('binary_file.bin', 'wb') as f:
    f.write(data)

print(" Binary file written")

# Reading binary data
print("\n 2ï¸ Reading Binary Data:")
print("-" * 60)

with open('binary_file.bin', 'rb') as f:
    binary_content = f.read()
    print(f"Read: {binary_content}")
    print(f"Type: {type(binary_content)}")
    print(f"Decoded: {binary_content.decode('ascii')}")

# Copying a binary file
print("\n 3ï¸ Copying Binary File:")
print("-" * 60)

with open('binary_file.bin', 'rb') as source:
    with open('binary_copy.bin', 'wb') as dest:
        dest.write(source.read())

print(" Binary file copied")

# Working with image files (example)
print("\n 4ï¸ Image File Example:")
print("-" * 60)
print("""
To copy an image:

with open('photo.jpg', 'rb') as source:
    with open('photo_copy.jpg', 'wb') as dest:
        dest.write(source.read())

Always use 'rb'/'wb' for images!
""")
````

## **10. Exception Handling with Files**

### **Common File Exceptions**

1. **`FileNotFoundError`**: File doesn't exist (reading)
2. **`PermissionError`**: No permission to access file
3. **`IOError`**: General I/O failure
4. **`IsADirectoryError`**: Tried to open directory as file
5. **`FileExistsError`**: File already exists (with 'x' mode)

### **Best Practices**

- Always use specific exception types
- Provide helpful error messages
- Use `with` statement for automatic cleanup
- Check file existence before operations

````python
print("Exception Handling Demo\n" + "="*60)

# 1. FileNotFoundError
print("\n 1ï¸ Handling FileNotFoundError:")
print("-" * 60)

try:
    with open('nonexistent.txt', 'r') as f:
        content = f.read()
except FileNotFoundError:
    print(" Error: File not found!")
    print("   Please check the filename and try again.")

# 2. PermissionError
print("\n 2ï¸ Handling PermissionError:")
print("-" * 60)

try:
    # This might fail on some systems
    with open('/root/protected.txt', 'w') as f:
        f.write('test')
except PermissionError:
    print(" Error: Permission denied!")
    print("   You don't have access to this location.")

# 3. Complete error handling
print("\n 3ï¸ Complete Error Handling:")
print("-" * 60)

def safe_read_file(filename):
    """
    Safely read a file with comprehensive error handling.
    """
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            return f.read()
    
    except FileNotFoundError:
        print(f"  File '{filename}' not found")
        return None
    
    except PermissionError:
        print(f" Permission denied for '{filename}'")
        return None
    
    except UnicodeDecodeError:
        print(f" Cannot decode '{filename}' (not a text file?)")
        return None
    
    except Exception as e:
        print(f" Unexpected error: {e}")
        return None

# Test the function
content = safe_read_file('example.txt')
if content:
    print(f" Read {len(content)} characters")

content = safe_read_file('missing.txt')
if content is None:
    print("   Handled gracefully!")

# 4. Using finally
print("\n 4ï¸ Using finally Block:")
print("-" * 60)

f = None
try:
    f = open('finally_demo.txt', 'w')
    f.write('test')
    # Simulate error
    # raise ValueError("Simulated error")
except Exception as e:
    print(f"Error: {e}")
finally:
    if f:
        f.close()
        print(" File closed in finally block")

print("\n 'with' statement handles this automatically!")
````

### **11.1 Working with CSV Files**

````python
import csv

print("CSV File Handling\n" + "="*60)

# Writing CSV
print("\n 1ï¸ Writing CSV File:")
print("-" * 60)

students = [
    ['Name', 'Age', 'Grade'],
    ['Alice', 20, 'A'],
    ['Bob', 22, 'B'],
    ['Charlie', 21, 'A']
]

with open('students.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(students)

print(" CSV file written")

# Reading CSV
print("\n 2ï¸ Reading CSV File:")
print("-" * 60)

with open('students.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)

# Using DictReader
print("\n 3ï¸ CSV as Dictionary:")
print("-" * 60)

with open('students.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"{row['Name']} - Grade: {row['Grade']}")
````

````python
import json

print("\nJSON File Handling\n" + "="*60)

# Writing JSON
print("\n 1ï¸ Writing JSON File:")
print("-" * 60)

data = {
    'name': 'Alice',
    'age': 25,
    'courses': ['Python', 'Data Science'],
    'active': True
}

with open('data.json', 'w') as f:
    json.dump(data, f, indent=2)

print("JSON file written")

# Reading JSON
print("\n 2ï¸ Reading JSON File:")
print("-" * 60)

with open('data.json', 'r') as f:
    loaded_data = json.load(f)

print(f"Name: {loaded_data['name']}")
print(f"Courses: {', '.join(loaded_data['courses'])}")
````

### **11.3 File Iteration with Chunks**

````python
print("\nReading Large Files in Chunks\n" + "="*60)

# Create a large file
with open('large_file.txt', 'w') as f:
    for i in range(1000):
        f.write(f"Line {i}: This is some data\n")

print(" Large file created (1000 lines)")

# Read in chunks
print("\nReading in 1KB chunks:")
print("-" * 60)

chunk_size = 1024  # 1KB
chunks_read = 0

with open('large_file.txt', 'r') as f:
    while True:
        chunk = f.read(chunk_size)
        if not chunk:
            break
        chunks_read += 1

print(f"Read {chunks_read} chunks")
print(" Memory-efficient for huge files!")
````

## **Summary & Best Practices**

### ** Best Practices Checklist**

1. **Always use `with` statement** for automatic file closure
2. **Specify encoding** explicitly (use `'utf-8'`)
3. **Handle exceptions** - don't let your program crash
4. **Use appropriate mode** - understand the difference!
5. **Check file existence** before operations
6. **Use pathlib** for modern path handling
7. **Read large files** line by line or in chunks
8. **Close files** if not using `with`
9. **Use binary mode** for non-text files
10. **Validate file operations** - check return values

### **Common Mistakes to Avoid**

- Forgetting to close files
- Using `'w'` when you mean `'a'` (data loss!)
- Reading huge files entirely into memory
- Not handling exceptions
- Using text mode for binary files
- Forgetting newlines in `write()`
- Not specifying encoding

### **Quick Reference**

```python
# Read entire file
with open('file.txt', 'r') as f:
    content = f.read()

# Read line by line (memory efficient)
with open('file.txt', 'r') as f:
    for line in f:
        print(line.strip())

# Write to file
with open('file.txt', 'w') as f:
    f.write('Hello\n')

# Append to file
with open('file.txt', 'a') as f:
    f.write('More data\n')

# Copy file
with open('source.txt', 'r') as src:
    with open('dest.txt', 'w') as dst:
        dst.write(src.read())
```
