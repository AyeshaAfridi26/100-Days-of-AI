**Creating ranges**

The range() function can be called with 1, 2, or 3 arguments, using this syntax:

range(start, stop, step)

Call range() With One Argument

If the range function is called with only one argument, the argument represents the stop value.

```python
x = range(10)
```

Call range() With Two Arguments

If the range function is called with two arguments, the first argument represents the start value, and the second argument represents the stop value.

```python
x = range(3, 10)
print(x)
```

Call range() With Three Arguments

If the range function is called with three arguments, the third argument represents the step value.

```python
x = range(3, 10, 2)
```

## Arrays
NumPy arrays are a part of the NumPy library, which is a powerful tool for numerical computing in Python



| Feature | Python Lists | NumPy Arrays |
| :--- | :--- | :--- |
| **Data Types** | Can hold different types (integers, strings, objects). | Must hold the same data type (homogenous/coerced). |
| **Performance** | Slower for large datasets; high memory overhead. | Significantly faster and more memory-efficient. |
| **Operations** | Operations (like `+`) concatenate or repeat lists. | Supports element-wise mathematical operations. |
| **Flexibility** | Built-in to Python; no imports required. | Requires importing a library (usually `numpy`). |

```python
import numpy as np
a = np.array([1, 2, 3, 4])

# Element-wise operations
print(a )

# Multi-dimensional array
res = np.array([[1, 2], [3, 4]])
print(res * 2)
```

### 1. Data Types (Heterogeneous vs. Homogenous)

```python
Lst: [1, "Data", 3.14] #(Keeps all types as they are)

Aray: np.array([1, "Data", 3.14]) #(Converts everything to strings: ['1', 'Data', '3.14'])
```

### 2. Performance (Memory and Speed)

Lists store a "pointer" for every single element, which is like having a map to find every item in a house. Arrays store items side-by-side in a single block of memory.

    Example: If you have 1 million numbers, a NumPy array will typically be 5x to 10x smaller in memory and much faster to process than a Python list.

```python
!pip install NumPy as np
```

```python
import sys
import numpy as np

size = 10000000
py_list = list(range(size))
np_arr = np.array(range(size), dtype=np.int64)

# Calculate true list size: size of container + (size of one int * number of ints)
one_int_size = sys.getsizeof(py_list[0])
total_list_size = sys.getsizeof(py_list) + (one_int_size * size)

print(f"True List Size:  {total_list_size / 1e6:.2f} MB")
print(f"NumPy Array Size: {np_arr.nbytes / 1e6:.2f} MB")
```

### 3. Operations (Concatenation vs. Math)

```python
list1 = [1, 2, 3]
print(list1 + list1)
# Result: [1, 2, 3, 1, 2, 3] (It just glued them together)
```

```python
arr1 = np.array([1, 2, 3])
print(arr1 + arr1)
# Result: [2, 4, 6] (It did actual addition)
```

    List: You can use them the second you install Python. No extra steps.

| Feature | Python List | array.array (Built-in) | NumPy Array (`ndarray`) |
| :--- | :--- | :--- | :--- |
| **Data Types** | Mixed (Int, Str, Obj) | Uniform (Single type code) | Uniform (Single `dtype`) |
| **Performance** | Slow (High overhead) | Faster (Memory efficient) | Fastest (C-level speed) |
| **Memory Layout** | Non-contiguous (Pointers) | Contiguous (Values) | Contiguous (Values) |
| **Dimensions** | Nested (List of lists) | 1D Only | N-Dimensional (Matrices) |
| **Vectorized Math**| ❌ No (Concatenates) | ❌ No | ✅ Yes (Element-wise) |
| **Import Needed?** | ❌ No | ✅ `import array` | ✅ `import numpy as np` |
| **Use Case** | General-purpose data | Saving memory on 1D data | Data Science & Math |

Vectorized math is a method of performing mathematical operations on an entire collection of data (an array) at once, rather than looping through individual elements one by one.  

In standard Python, if you want to double 1,000,000 numbers, you have to tell the computer to "get the first number, multiply it, get the second number, multiply it..." and so on. In vectorized math, you simply say "multiply this whole block by 2."

```python
import numpy as np
import array

# List: Keeps mixed types
py_list = [10, "Python", 3.14]
print(f"List Types: {[type(item) for item in py_list]}")

# Built-in Array: Will ERROR if you try to mix types
try:
    arr_built_in = array.array('i', [10, "Python"])
except TypeError:
    print("Built-in Array: Error! Cannot mix strings and integers.")

# NumPy: Coerces everything to a string (Unicode)
np_arr = np.array([10, "Python", 3.14])
print(f"NumPy Type: {np_arr.dtype} (Everything became a string)")
```

```python
# List: Multiplication repeats the list
print(f"List [1,2,3] * 2: {[1, 2, 3] * 2}")

# Built-in Array: Multiplication also repeats (like a list)
arr_built_in = array.array('i', [1, 2, 3])
print(f"Built-in Array * 2: {arr_built_in * 2}")

# NumPy: Multiplication applies to every NUMBER
np_arr = np.array([1, 2, 3])
print(f"NumPy Array * 2: {np_arr * 2}")
```

```python
import sys

size = 1000000 # 1 Million items
py_list = list(range(size))
np_arr = np.arange(size, dtype=np.int64)

# List size = (Pointer container) + (Actual Int objects)
list_total = sys.getsizeof(py_list) + (sys.getsizeof(py_list[0]) * size)
# NumPy size = Total memory used by the block
numpy_total = np_arr.nbytes

print(f"List True Size: {list_total / 1e6:.2f} MB")
print(f"NumPy True Size: {numpy_total / 1e6:.2f} MB")
```

```python
# List: Nested list (List of lists)
list_2d = [[1, 2], [3, 4]]
print(f"List element access: {list_2d[0][1]}") # Requires two separate lookups

# Built-in Array: ONLY supports 1D
try:
    bad_arr = array.array('i', [[1, 2], [3, 4]])
except TypeError:
    print("Built-in Array: Cannot handle 2D structures.")

# NumPy: True 2D Matrix
np_2d = np.array([[1, 2], [3, 4]])
print(f"NumPy access [row, col]: {np_2d[0, 1]}") # Direct matrix coordinate
```

### Adding Elements to an Array

Elements can be added to the Python Array by using built-in insert() function.

append() is also used to add the value mentioned in its arguments at the end of the Python array.

```python
import array as arr
a = arr.array('i', [1, 2, 3])
print(*a)

a.insert(1, 4)  # Insert 4 at index 1
print(*a)
```

# Python Iterators

An iterator is an object that contains a countable number of values.

An iterator is an object that can be iterated upon, meaning that you can traverse through all the values.

Iterator vs Iterable

Lists, tuples, dictionaries, and sets are all iterable objects. They are iterable containers which you can get an iterator from.

All these objects have a iter() method which is used to get an iterator:

```python
mytuple = ("apple", "banana", "cherry")
myit = iter(mytuple)

print(next(myit))
print(next(myit))
print(next(myit))
```

```python
# Strings are also iterable objects, containing a sequence of characters

mystr = "banana"
myit = iter(mystr)

print(next(myit))
print(next(myit))
print(next(myit))
print(next(myit))
print(next(myit))
print(next(myit))
```

Looping Through an Iterator

We can also use a for loop to iterate through an iterable object:

```python
mytuple = ("apple", "banana", "cherry")

for x in mytuple:
  print(x)
```

## Python Dates

A date in Python is not a data type of its own, but we can import a module named datetime to work with dates as date objects.

```python
import datetime

x = datetime.datetime.now()
print(x)
```

The datetime module has many methods to return information about the date object.

```python
import datetime

x = datetime.datetime.now()

print(x.year)
print(x.strftime("%B")) # Return the year and name of weekday:
```

Creating Date Objects

To create a date, we can use the datetime() class (constructor) of the datetime module.

The datetime() class requires three parameters to create a date: year, month, day.

```python
import datetime

x = datetime.datetime(2020, 5, 17)

print(x)
```

The strftime() Method

The datetime object has a method for formatting date objects into readable strings.

The method is called strftime(), and takes one parameter, format, to specify the format of the returned string:

```python
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%B"))
```

# Python Math

Python has a set of built-in math functions, including an extensive math module, that allows you to perform mathematical tasks on numbers.

## Built-in Math Functions

The min() and max() functions can be used to find the lowest or highest value in an iterable:

```python
x = min(5, 10, 25)
y = max(5, 10, 25)

print(x)
print(y)
```

The abs() function returns the absolute (positive) value of the specified number

```python
x = abs(-7.25)

print(x)
```

The pow(x, y) function returns the value of x to the power of y (x^y).

```python
x = pow(4, 3)

print(x)
```

## The Math Module

Python has also a built-in module called math, which extends the list of mathematical functions.

To use it, you must import the math module:

```python
# The math.sqrt() method for example, returns the square root of a number:

import math

x = math.sqrt(25)

print(x)
```

The math.ceil() method rounds a number upwards to its nearest integer, and the math.floor() method rounds a number downwards to its nearest integer, and returns the result:

```python
import math

x = math.ceil(1.1)
y = math.floor(1.9)

print(x) # returns 2
print(y) # returns 1
```

```python
# The math.pi constant, returns the value of PI (3.14...):
# Example
import math

x = math.pi

print(x)
```
