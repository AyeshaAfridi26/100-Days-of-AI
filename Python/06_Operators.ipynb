### Boolean Values
In programming you often need to know if an expression is True or False.
When you compare two values, the expression is evaluated and Python returns the Boolean answer

```python
print(10 > 9)
print(10 == 9)
print(10 < 9)
```

```python
a = 200
b = 33

if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")
```

Almost any value is evaluated to True if it has some sort of content.

Any string is True, except empty strings.

Any number is True, except 0.

Any list, tuple, set, and dictionary are True, except empty ones
there are not many values that evaluate to False, except empty values, such as (), [], {}, "", the number 0

```python
#Almost any value is evaluated to True if it has some sort of content.
print(bool("Hello"))
print(bool(15))
print(bool())
```

```python
print(bool(False))
print(bool(None))
print(bool(0))
print(bool(""))
print(bool(()))
print(bool([]))
print(bool({}))
```

Arithmetic Operators

Arithmetic operators are used with numeric values to perform common mathematical operations:
| Operator | Name | Example | Result (if x = 15, y = 4) |
| :---: | :--- | :--- | :--- |
| `+` | Addition | `x + y` | `19` |
| `-` | Subtraction | `x - y` | `11` |
| `*` | Multiplication | `x * y` | `60` |
| `/` | Division | `x / y` | `3.75` |
| `%` | Modulus (Remainder) | `x % y` | `3` |
| `**` | Exponentiation (Power) | `x ** y` | `50625` |
| `//` | Floor division | `x // y` | `3` |

```python
x = 15
y = 4

print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x % y)
print(x ** y)
print(x // y)
```

 Assignment Operators

Assignment operators are used to assign values to variables: 

| Operator | Example | Same As | Description |
| :---: | :--- | :--- | :--- |
| `=` | `x = 5` | `x = 5` | Assigns a value to a variable |
| `\+=` | `x += 3` | `x = x + 3` | Addition assignment |
| `\-=` | `x -= 3` | `x = x - 3` | Subtraction assignment |
| `*=` | `x *= 3` | `x = x * 3` | Multiplication assignment |
| `/=` | `x /= 3` | `x = x / 3` | Division assignment |
| `%=` | `x %= 3` | `x = x % 3` | Modulus assignment (remainder) |
| `//=` | `x //= 3` | `x = x // 3` | Floor division assignment |
| `**=` | `x **= 3` | `x = x ** 3` | Exponentiation assignment |
| `&=` | `x &= 3` | `x = x & 3` | Bitwise AND assignment |
| `\|=` | `x \|= 3` | `x = x \| 3` | Bitwise OR assignment |
| `^=` | `x ^= 3` | `x = x ^ 3` | Bitwise XOR assignment |
| `>>=` | `x >>= 3` | `x = x >> 3` | Bitwise right shift assignment |
| `<<=` | `x <<= 3` | `x = x << 3` | Bitwise left shift assignment |
| `:=` | `print(x := 3)` | `x = 3 <br> print(x)` | Walrus operator (Assigns and returns) |

```python
numbers = [1, 2, 3, 4, 5]
count = len(numbers)
if count > 3:
    print(f"List has {count} elements")

if (count := len(numbers)) > 3:
    print(f"List has {count} elements")
```

Comparison Operators

Comparison operators are used to compare two values:

| Operator | Name | Example | Result (if x = 5, y = 3) |
| :---: | :--- | :--- | :--- |
| `==` | Equal | `x == y` | `False` |
| `!=` | Not equal | `x != y` | `True` |
| `>` | Greater than | `x > y` | `True` |
| `<` | Less than | `x < y` | `False` |
| `>=` | Greater than or equal to | `x >= y` | `True` |
| `<=` | Less than or equal to | `x <= y` | `False` |

```python
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print(x is z)
print(x is y)
print(x == y)
```

```python
x = ["apple", "banana"]
y = ["apple", "banana"]
```

Difference Between is and ==

    is - Checks if both variables point to the same object in memory
    == - Checks if the values of both variables are equal


```python
x = [1, 2, 3]
y = [1, 2, 3]

print(x == y)
print(x is y)
```

### Membership Operators

Membership operators are used to test if a sequence is presented in an object:
Operator 	Description 	Example 	Try it
in  	Returns True if a sequence with the specified value is present in the object 	x in y
not in 	Returns True if a sequence with the specified value is not present in the object 	x not in y

```python
fruits = ["apple", "banana", "cherry"]

print("banana" in fruits)
```

```python
fruits = ["apple", "banana", "cherry"]

print("pineapple" not in fruits)
```

```python
text = "Hello World"

print("H" in text)
print("hello" in text)
print("z" not in text)
```
