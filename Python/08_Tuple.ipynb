# Tuples
A tuple in Python is an immutable ordered collection of elements.

    Tuples are similar to lists, but unlike lists, they cannot be changed after their creation (i.e., they are immutable).
    Tuples can hold elements of different data types.
    The main characteristics of tuples are being ordered, heterogeneous and immutable

```python
tup = ("A",1,1.2,False)
print(tup)

# Using String
tup = ('Geeks', 'For')
print(tup)
# listt = list((1,2,3,4))
# Using List
li = [1, 2, 4, 5, 6]
print(tuple(li))

# Using Built-in Function
tup = tuple('Geeks')
print(tup)
```

```python
tup = (5, 'Welcome', 7, 'Geeks')
print(tup)

# Creating a Tuple with nested tuples
tup1 = (0, 1, 2, 3)
tup2 = ('python', 'geek')
tup3 = (tup1, tup2)
print(tup3)

# Creating a Tuple with repetition
tup1 = ('Geeks',) * 3
print(tup1)
lst = [2]*5 # [2,2,2,2,2]

# Creating a Tuple with the use of loop
tup = ('Geeks')
n = 5 #Hard coded
for i in range(int(n)):
    tup = (tup,)
    print(tup)
```

Python Tuple Basic Operations

Below are the Python tuple operations.

    Accessing of Python Tuples
    Concatenation of Tuples
    Slicing of Tuple
    Deleting a Tuple

```python
# Accessing Tuple with Indexing
tup = tuple("Geeks")
print(tup[0])

# Accessing a range of elements using slicing
print(tup[1:4])
print(tup[:3])

# Tuple unpacking
tup = ("Geeks", "For", "Geeks")

# This line unpack values of Tuple1
a, b, c = tup
print(a)
print(b)
print(c)
```

```python
tup = (1,2,3,4)
print("Before slicing",id(tup))
tup = tup[1:2] # creates a new tuple
print("after slicing",id(tup))
```

```python
tup[0]=2 # Tuple objects do not support item assignment
```

```python
tup = (1,2,3,4)
old_id = id(tup)

tup = tup[1:2]
new_id = id(tup) #object’s memory address.

print(old_id, new_id)
print(old_id == new_id)
```

```python
# a list (mutable)
inner = [1, 2, 3]

# a tuple (immutable container) holding that list
t = (inner,)

print("id(inner):", id(inner))
print("id(t[0]) :", id(t[0]))   # same object as inner
print("Same object?", inner is t[0])

# mutate the list
t[0].append(99)

print("After append:", t)
```
