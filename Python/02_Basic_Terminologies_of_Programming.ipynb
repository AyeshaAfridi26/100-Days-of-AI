## Low vs High Level Language
###Low-Level Language:
Low-level languages are the "native tongue" of the computer.
When you write at this level, you aren't just giving math instructions; you are literally moving electrical signals between specific physical parts of the processor. For example:
  - Machine Code: The absolute lowest level. It consists of sequences of binary (0s and 1s) that the CPU executes directly.
  - Assembly Language: One step above machine code. It uses "mnemonics" (short words) instead of numbers, but it still maps 1-to-1 with CPU instructions.

### High-Level Language:
High-level languages are designed to be human-readable. They use English-like words and abstract away the complex hardware details. You don't have to worry about memory addresses or CPU registers; the language handles that for you behind the scenes. Examples:
  - Python, Java, JavaScript, C++.  

## Compiled Languages vs Interpreted Languages
### Compiled Languages (The "Pre-Translated" Recipe)

In a compiled language, the computer takes your entire source code ( the code you have written ) and translates it all at once into Machine Code (binary - Language of 0's and 1's) before you ever run the program. This "translation" creates a new, separate file (like an .exe on Windows).

    How it works: You write the code → Run the Compiler → It creates an Executable → You run the Executable.

    The Pro: It is incredibly fast because the computer doesn't have to "read" and "think" while running; it just follows the pre-translated binary instructions.

    The Con: If you change one single letter of code, you have to re-translate (re-compile) the whole thing.

### Interpreted Languages (The "Live Translator")

An interpreted language doesn't translate the code beforehand. Instead, a program called the Interpreter reads your code line-by-line and executes it on the fly.

    How it works: You write the code → You run the code directly → The Interpreter translates each line as it reaches it.

    The Pro: It’s much faster to develop (entire process of building the software). You can write code and test it instantly without waiting for a "build" or "compile" step.

    The Con: It is generally slower to execute(also called "running") because the computer is doing two jobs at once: translating and running.    
    
  Python is a bit of a hybrid but mainly considered as an interpreted language. When you "translate" Python, it actually goes through a two-step translation:

    Source Code → Bytecode (A halfway point that's easier for machines to read).

    Bytecode → Machine Code (The final translation done by the Python Virtual Machine).

| Feature         | Compiled (e.g., C++)              | Interpreted (e.g., Python)                  |
| --------------- | --------------------------------- | ------------------------------------------- |
| Translation     | Entire program at once.           | Line by line, during execution.             |
| Output File     | A standalone `.exe` or binary.    | No standalone file (needs the interpreter). |
| Execution Speed | Very fast.                        | Slower (due to overhead).                   |
| Error Finding   | Found all at once before running. | Found only when the line is reached.        |
| Best For        | Games, Operating Systems.         | Data Science, Web, Scripting.               |

## Programming Terminologies used in Python
### Variable:
A variable is a reserved memory location to store values. In Python, you don't need to declare what kind of data you are storing (like numbers or text); you just give it a name and an assignment.

    Technical Detail: Python is **dynamically typed**, meaning a variable name can point to a number at one moment and a string the next.

### DataTypes:
Data types define what kind of data a variable holds. This tells the computer how much memory to use and what operations are allowed.

###Syntax (The "Grammar")

Syntax refers to the set of rules that defines how a Python program is written and interpreted. If you break a syntax rule, the code won't run at all.

    Technical Detail: Python relies heavily on Indentation (whitespace) to define blocks of code, whereas other languages use curly braces {}.

### Functions (The "Recipes")

A function is a reusable block of code that only runs when it is called. It takes inputs, performs a task, and returns a result.

    Technical Detail: This follows the DRY principle (Don't Repeat Yourself).

    Syntax: Use the def keyword.
### Comments (The "Notes")

Comments are lines of text that the computer ignores during execution. They are strictly for humans to explain what the code is doing.

    Syntax: Use the # symbol for single lines.
###Expressions vs. Statements

    Expression: A piece of code that produces a value (e.g., 5 + 5).

    Statement: A complete action or instruction (e.g., x = 10 or a print() call).

### Keywords (The "Reserved Words")

Keywords are special words that Python has already "claimed" for its own use. You cannot use these as names for your variables because Python uses them to understand the structure of your code.

    Examples: if, else, while, for, def, class, import.

    Technical Detail: There are about 35 keywords in Python. They are case-sensitive (e.g., True is a keyword, but true is not).
### Operators (The "Action Symbols")

Operators are symbols that perform operations on variables and values.

    Arithmetic Operators: Used for math: + (add), - (subtract), * (multiply), / (divide).

    Assignment Operators: Used to give a value to a variable: = (not to be confused with "equal to").

    Comparison Operators: Used to check relationships: == (equal to), != (not equal to), > (greater than).

### Literals (The "Raw Data")

A Literal is a fixed value written directly in the source code. While a variable is a name for a value, a literal is the value itself.

    String Literal: "Hello World"

    Integer Literal: 42

    Float Literal: 3.14

### Input and Output (I/O)

This is how your program talks to the outside world.

    Output: Sending data to the screen using the print() function.

    Input: Getting data from the user (usually via the keyboard) using the input() function.

### Objects

In Python, everything is an object. Whether it is a number, a string, or a complex library, Python treats it as an "object" that has:

    Identity: Where it is in memory.

    Type: What kind of thing it is (e.g., a string).

    Value: The data it holds (e.g., "Hello").

### Flow Control (The "Decision Maker")

Flow Control refers to the order in which individual statements are executed. By default, a program runs from top to bottom, but flow control lets you skip sections or repeat them.

    Conditionals (if, elif, else): These allow the program to make decisions based on whether something is True or False.

    Loops (for, while): These allow you to repeat a block of code multiple times.

### Arguments and Parameters (The "Inputs")

These terms are often used interchangeably, but there is a slight technical difference when working with Functions.

    Parameter: The variable listed in the function definition (the "placeholder").

    Argument: The actual value you pass into the function when you call it.

    def greet(name): # 'name' is the Parameter
      print("Hello " + name)
    greet("Alice")   # "Alice" is the Argument    


### Scope (The "Visibility")

Scope determines where a variable can be seen or used within your code.

    Global Scope: A variable defined outside of any function. It can be accessed from anywhere in the file.

    Local Scope: A variable defined inside a function

### Modules and Packages (The "Organizers")

As your code gets bigger, you don't want it all in one file.

    Module: A single Python file (.py) containing code you want to reuse.

    Package: A folder containing multiple Python modules.

    Import: The keyword used to bring a module or package into your current file.

### Bugs and Debugging

    Bug: An error in the code that causes it to crash or produce the wrong result.

    Debugging: The process of finding and fixing those bugs.

    Traceback: The error message Python prints when something goes wrong. It shows you exactly which line caused the "crash."

### Data Structures (The "Organizers")

Beyond simple variables, you need ways to store collections of data.

    List: An ordered collection of items (like a shopping list). Syntax: [1, 2, 3]

    Dictionary (Dict): A collection of "Key-Value" pairs (like a real dictionary where you look up a word to find a definition). Syntax: {"name": "Gemini", "age": 1}


### Exceptions (The "Runtime Errors")

An Exception is an error that happens while the program is executing. Even if your syntax is perfect, something might go wrong (like trying to divide some number by zero).

### Immutable vs. Mutable (The "Changeability")

This is a very important Python concept regarding how data is stored in memory.

    Mutable: The data can be changed after it is created. (Example: Lists—you can add or remove items).

    Immutable: The data cannot be changed. If you want to change it, Python has to create a brand new object. (Example: Strings and Tuples).

### IDE (Integrated Development Environment)

An IDE is the software you use to write your code. It’s like a "Word Processor for Programmers." It provides tools like:

    Syntax Highlighting: Coloring keywords to make them easier to read.

    Auto-complete: Suggesting code as you type.

    Debugger: A tool to pause your code and see what’s happening inside.

    Examples: PyCharm, VS Code, or IDLE (which comes with Python).

### Garbage Collection (The "Auto-Clean")

In low-level languages, you have to manually delete data from the computer's memory when you're done with it. If you forget, the computer slows down. Python has a Garbage Collector.

    What it is: A background process that automatically identifies objects that are no longer being used by your program and deletes them to free up RAM.

### Pointers vs. References

While Python hides the actual "address" of your data, it's important to know how it handles variables.

    Pointer (Low Level): A variable that stores the physical memory address (like 0x7ffe) of another value.

    Reference (Python): Python variables are references. When you say x = y, you aren't creating a second copy of the data; you are just making two names point to the same "object" in memory.

### Iterables and Iterators

This is the terminology for things you can "loop" over.

    Iterable: Any object that can return its members one at a time (like a List, a String, or a Tuple).

    Iterator: The actual "engine" that keeps track of where you are in that list while you are looping.

    Analogy: An Iterable is a book. An Iterator is your finger pointing at the current word as you read.

                    


### Assignment (The =)

In math, = means "these two things are equal." In programming, it means "take the value on the right and STUFF it into the variable on the left."

    Assignment: x = 10 (Put 10 in x)

    Re-assignment: You can change it later. x = 20 (The 10 is gone, x is now 20).

    Augmented Assignment: Shortcuts for math.

        x += 5 is the same as x = x + 5

### Ordered vs. Unordered Collections

When you store a group of items, how they are stored in memory matters.

    Ordered: The items stay in the exact position you put them in. They have an Index (a number representing their position, starting at 0).

        Example: Lists ["Apple", "Banana"]. Apple is always at 0.

    Unordered: The items are just "in a bag." There is no first or last. You find things by a unique name (a Key), not a position.

        Example: Dictionaries {"fruit": "Apple"} or Sets.
### Function vs. Method (How to tell them apart)

This is a very common point of confusion. Both are blocks of code that "do something," but their ownership is different.

Function (Independent)

A function is a standalone "recipe." You call it by its name and pass data into it.

    Syntax: function_name(data)

    Example: print("Hello") or len([1,2,3])

Method (Dependent/Owned)

A method is a function that belongs to an object. It is "attached" to a specific piece of data.

    Syntax: data.method_name()

    Example: "hello".upper() (The .upper() belongs to the string).

    Pro-Tip: If there is a dot . before the name, it’s a Method.    
### Return Values (The "Output" of a Function)

When a function finishes its job, it can either just "do something" (like printing to the screen) or it can hand back a result to the main program.

    No Return: Like a waiter showing you a picture of food. You see it, but you can't use it.

    Return: Like a waiter giving you the actual plate of food. You can now store it in a variable and use it later.
### Comparison vs. Identity

Earlier we talked about assignment (=). Now look at how we check things:

    == (Comparison): Checks if the Value is the same. (Are these two boxes holding the same thing?)

    is (Identity): Checks if they are the Exact Same Object in memory. (Are these two names referring to the exact same physical box?)
### Concatenation

How do you join two strings together?

    Concatenation: Using the + sign to glue strings. (e.g., "Hi " + "there").            

### Indexing and Slicing (Finding Data)

Since lists and strings are Ordered, we use numbers to find specific parts.

    Index: The position of a single item. Note: Python starts counting at 0!

    Slice: Grabbing a "chunk" or range of items. Syntax: [start:stop].
