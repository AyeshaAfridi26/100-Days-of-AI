## What is Matplotlib?

Matplotlib is a Python library used for drawing charts and graphs from data. You hand it some numbers and it gives you back a picture: line plots, bar charts, scatter plots, histograms, and so on. It was first released in 2003 and has been the default plotting library in Python ever since.

A lot of other plotting tools are actually built on top of it:
- pandas' `.plot()` method calls Matplotlib under the hood
- seaborn is a thin layer over Matplotlib that handles styling
- Plotly and Bokeh are alternatives, but Matplotlib is what you'll see in most tutorials and Stack Overflow answers

If you learn Matplotlib first, you can read and modify almost any Python plot you come across in the wild.

### What it's good at

- Very flexible. You can control nearly every pixel if you want to.
- Runs everywhere — Jupyter notebooks, plain Python scripts, web apps.
- Big community, so answers to most problems are one search away.
- Works smoothly with NumPy and pandas.

### Where it falls short

- The API feels a little dated in places and isn't always consistent.
- The defaults aren't the prettiest right out of the box.
- It's not really built for interactive dashboards. Reach for Plotly when you need clicking, zooming, or hovering.

### What you'll actually use it for

- Looking at data while you're exploring it
- Final figures for a report, paper, or presentation
- Tracking ML training — loss curves, accuracy plots, confusion matrices
- Statistical plots like distributions and correlation views

## Setup and Import

The standard import line is `import matplotlib.pyplot as plt`. Almost every Matplotlib example you see online starts this way.

A quick note on naming: **pyplot** is a sub-module inside Matplotlib that holds the high-level plotting functions — the ones you actually call to draw things. The rest of Matplotlib is the lower-level machinery sitting behind it. When you see `plt.plot(...)` or `plt.title(...)`, that `plt` is just the alias for `pyplot`.

````python
import matplotlib
matplotlib.use('Agg')   # Use a non-GUI backend (works everywhere — saves to files)
import matplotlib.pyplot as plt
import numpy as np

print("Matplotlib version:", matplotlib.__version__)
````

**On real Jupyter setups:** you don't need the `matplotlib.use('Agg')` line — just use `%matplotlib inline` at the top of your notebook. We're using `Agg` here so plots save without needing a screen.

## The Two APIs — pyplot vs Object-Oriented

This is probably the biggest source of confusion for people learning Matplotlib. There are two different ways to write the same plot, and you'll see both.

Before going further: an **API** (Application Programming Interface) just means the set of functions and patterns a library gives you to interact with it. Same library, two different styles for talking to it.

### Approach 1: pyplot (the "implicit" or MATLAB-style API)

In this style you call functions directly on `plt`. Matplotlib quietly keeps track of which figure and which plot is "current" in the background, so you don't have to. Quick to type, but it gets messy as plots grow.

````python
# Quick — fine for one-off plots
x = np.linspace(0, 10, 100)
print(x)
y = np.sin(x)
print(y)
plt.figure(figsize=(8, 4)) # 8 inches wide, 4 inches tall
plt.plot(x, y)
plt.title("Sine wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.savefig('mpl_01.png', dpi=80)
plt.close()
print("Saved mpl_01.png — using pyplot API")
````

### Approach 2: object-oriented (the "explicit" API) — recommended

Here you grab the Figure and Axes objects yourself and call methods on them. A bit more typing, but the code is much clearer about what it's doing to which plot.

````python
# More verbose but much clearer for complex plots
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 4))      # explicit fig and axes
ax.plot(x, y)
ax.set_title("Sine wave")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
fig.savefig('mpl_02.png', dpi=80) # 80 pixels per inch
plt.close()
print("Saved mpl_02.png — using object-oriented API")
````

Both styles produce the exact same plot. The only real difference is how the code reads:

| pyplot | Object-Oriented |
|---|---|
| `plt.plot(x, y)` | `ax.plot(x, y)` |
| `plt.title(...)` | `ax.set_title(...)` |
| `plt.xlabel(...)` | `ax.set_xlabel(...)` |
| `plt.savefig(...)` | `fig.savefig(...)` |

### Which one should you use?

Use the object-oriented (OO) style. Reasons:

- Once you have more than one subplot, the OO style is much less ambiguous — every call says exactly which Axes it's modifying.
- It's the style used in pandas, seaborn, and most serious tutorials, so you'll need to read it anyway.
- It scales: a 3-line throwaway plot looks fine either way, but a 50-line figure for a report is far easier to read in OO style.

The pyplot style is fine for a quick three-line check while you're exploring. For anything you'll keep around, prefer OO.

## Figure and Axes — the Core Mental Model

If you only take one thing from this notebook, make it this part. Every Matplotlib plot is built from two kinds of object:

| Object | What it is |
|---|---|
| **Figure** (`fig`) | The whole canvas. Think of it as a sheet of paper or a single slide. |
| **Axes** (`ax`) | One chart sitting on that canvas, with its own x-axis, y-axis, and the data drawn inside it. |

A Figure can hold a single Axes or many of them arranged in a grid.

One thing that throws people off: in Matplotlib, the word **Axes** does *not* mean "axis" in the math sense. An Axes is the entire plotting area for one chart — the box, the tick marks, the gridlines, the data, everything. The x-axis and y-axis are inside the Axes. Confusing for about a day, then it sticks.

```
+---------------------------- Figure ----------------------------+
|                                                                |
|   +--- Axes ---+    +--- Axes ---+    +--- Axes ---+         |
|   |            |    |            |    |            |         |
|   |  plot 1    |    |  plot 2    |    |  plot 3    |         |
|   |            |    |            |    |            |         |
|   +------------+    +------------+    +------------+         |
|                                                                |
+----------------------------------------------------------------+
```

### Creating a figure with multiple subplots

The function `plt.subplots(rows, cols)` is how you ask for a Figure with a grid of Axes inside it. The word **subplot** just means "one of the smaller charts inside the bigger figure".

````python
# A figure with 2 axes side by side
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

x = np.linspace(0, 10, 100)

# axes[0] is the LEFT axes
axes[0].plot(x, np.sin(x))
axes[0].set_title("sin(x)")

# axes[1] is the RIGHT axes
axes[1].plot(x, np.cos(x), color='red')
axes[1].set_title("cos(x)")

fig.suptitle("Two subplots")
plt.tight_layout() #  automatically adjusts the spacing around and between subplots 
fig.savefig('mpl_03.png', dpi=80)
plt.show()
plt.close()
print("Saved mpl_03.png — two side-by-side plots")
````

````python
# A 2x2 grid
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))

x = np.linspace(0, 10, 100)

axes[0, 0].plot(x, np.sin(x));        axes[0, 0].set_title("sin(x)")
axes[0, 1].plot(x, np.cos(x));        axes[0, 1].set_title("cos(x)")
axes[1, 0].plot(x, x**2);             axes[1, 0].set_title("x²")
axes[1, 1].plot(x, np.exp(x/3));      axes[1, 1].set_title("eˣᐟ³")

plt.tight_layout()
fig.savefig('mpl_04.png', dpi=80)
plt.close()
print("Saved mpl_04.png — 2x2 grid")
````

A bit more on what `plt.subplots(rows, cols)` actually hands back:

- One Figure object (the canvas)
- A NumPy array of Axes objects (the individual plots)

For a grid you index the array like `axes[row, col]`. For a single row or column it's `axes[i]`. If you call `plt.subplots()` with no arguments at all, you get one Figure and one Axes — the most common starting point.

## Your First Real Plots

A quick tour of the four most common plot types so you actually have something on the screen. Notebook 2 goes into each one in more depth — this is just enough to get plotting.

Short definitions before the code:

- **Line plot**: connects data points with a line. Good for showing how something changes over a continuous range, like time.
- **Scatter plot**: draws each data point as a separate dot. Good for showing the relationship between two variables.
- **Bar chart**: one bar per category, with height showing the value. Good for comparing groups.
- **Histogram**: takes a bunch of numbers, groups them into ranges (called *bins*), and shows how many fall into each. Good for seeing the shape of a distribution.

````python
# Line plot
fig, ax = plt.subplots(figsize=(8, 4))
x = np.linspace(0, 10, 50)
ax.plot(x, np.sin(x), label='sin')
ax.plot(x, np.cos(x), label='cos')
ax.set_title("Sin and Cos")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
ax.grid(True, alpha=0.3)
fig.savefig('mpl_05_line.png', dpi=80)
plt.close()
print("Saved line plot")
````

````python
# Scatter plot
np.random.seed(0)
fig, ax = plt.subplots(figsize=(7, 5))
x = np.random.randn(100)
y = 2 * x + np.random.randn(100)
ax.scatter(x, y, alpha=0.6)
ax.set_title("Scatter: y ≈ 2x + noise")
ax.set_xlabel("x")
ax.set_ylabel("y")
fig.savefig('mpl_05_scatter.png', dpi=80)
plt.close()
print("Saved scatter plot")
````

````python
# Bar chart
fig, ax = plt.subplots(figsize=(7, 4))
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 32, 67]
ax.bar(categories, values, color='steelblue')
ax.set_title("Sample Bar Chart")
ax.set_ylabel("Value")
fig.savefig('mpl_05_bar.png', dpi=80)
plt.close()
print("Saved bar chart")
````

````python
# Histogram
np.random.seed(0)
fig, ax = plt.subplots(figsize=(7, 4))
data = np.random.normal(100, 15, 1000)
ax.hist(data, bins=30, color='steelblue', edgecolor='black')
ax.set_title("Histogram — normal distribution")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")
fig.savefig('mpl_05_hist.png', dpi=80)
plt.close()
print("Saved histogram")
````

## Titles, Labels, Legends — the Essentials

A chart with no labels doesn't really communicate anything. Three pieces almost every plot should have:

- **Title** — a short sentence at the top saying what the chart is about.
- **Axis labels** — what the x-axis and y-axis actually represent, with units when it matters (seconds, dollars, count, etc.).
- **Legend** — a small box that matches each line or color to what it represents. You only need one when there's more than one thing plotted on the same Axes.

A **grid** (those faint background lines) is optional, but it makes values much easier to read off the chart.

````python
fig, ax = plt.subplots(figsize=(9, 5))

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), label='sin(x)', color='blue', linewidth=2)
ax.plot(x, np.cos(x), label='cos(x)', color='red', linewidth=2, linestyle='--')

# Always include these
ax.set_title("Wave functions — labeled properly", fontsize=14, fontweight='bold')
ax.set_xlabel("x (radians)", fontsize=12)
ax.set_ylabel("y", fontsize=12)

# Legend — pulls from the 'label=' arguments above
ax.legend(loc='upper right', fontsize=11)

# Grid for readability
ax.grid(True, alpha=0.3)

# Axes limits (optional — Matplotlib picks reasonable defaults)
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

fig.savefig('mpl_06.png', dpi=80)
plt.close()
print("Saved labeled plot")
````

### Legend positions

The `loc` argument accepts:
- `'best'` (default — tries to avoid data)
- `'upper right'`, `'upper left'`, `'lower right'`, `'lower left'`
- `'right'`, `'center left'`, `'center right'`, `'center'`
- A `(x, y)` tuple of figure coordinates

## Saving Figures

To save a chart to a file, call `fig.savefig()`. The most useful arguments:

- `dpi=` — short for *dots per inch*, controls resolution. The default of 100 is fine on screen; bump to 300 for anything you're printing.
- `bbox_inches='tight'` — trims off extra white space around the chart so it doesn't look like it's floating in the middle of a big margin.
- `transparent=True` — makes the background see-through. Useful for slides where you don't want a white block behind the chart.
- `facecolor='white'` — does the opposite: forces a solid white background regardless of theme.

### File formats

Matplotlib picks the format based on the file extension you give it:

| Extension | Best for |
|---|---|
| `.png` | The default choice. Lossless raster image (made of pixels). Works everywhere. |
| `.pdf` | Vector format (made of shapes, not pixels). Stays sharp at any zoom level. Best for papers and reports. |
| `.svg` | Vector too, but designed for the web. Editable in tools like Illustrator or Inkscape. |
| `.jpg` | Smaller files, but lossy compression makes lines and text look fuzzy. Avoid for charts. |

A note on raster vs vector: raster images store a fixed grid of colored pixels, so they get blurry when you zoom in. Vector images store the underlying shapes and lines, so they stay crisp at any size.

````python
# Save the same plot in multiple formats
fig, ax = plt.subplots(figsize=(7, 4))
x = np.linspace(0, 2*np.pi, 100)
ax.plot(x, np.sin(x))
ax.set_title("Save in different formats")

fig.savefig('output.png', dpi=150, bbox_inches='tight')
fig.savefig('output.pdf', bbox_inches='tight')
fig.savefig('output.svg', bbox_inches='tight')
plt.close()

import os
for fmt in ['png', 'pdf', 'svg']:
    size = os.path.getsize(f'output.{fmt}')
    print(f"output.{fmt}: {size:,} bytes")
````
